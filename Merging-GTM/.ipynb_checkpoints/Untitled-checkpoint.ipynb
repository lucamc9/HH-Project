```python
import pandas as pd
import numpy as np

gem = pd.read_csv('matrix.tbl', delimiter='\t', header=None)
gtm = pd.read_csv('test.txt', delimiter='\t', header=None)
```

```python
gem = gem[[0, 1, 2]]
```

```python
merged_df = pd.merge(gem, gtm, on=0, how='left').dropna(axis=0, how='any') # merge & make sure there's no NaN values
```

```python
term_df = merged_df.select_dtypes(include=[np.float])
```

```python
expr_df = merged_df.select_dtypes(include=[np.integer])
```

```python
term_df.iloc[:,0]
```

```python
final_df = merged_df[[0]]   # Initialize df with gene_names
for i in range(len(expr_df.columns)):
    final_df[i] = list(zip(expr_df.iloc[:,i], term_df.iloc[:,i].astype(int)))
```
